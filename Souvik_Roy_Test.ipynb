# **FINANCIAL PERFORMANCE ANALYSIS**

```python
Total_sales= df["Sales"].sum()
Total_sales
```

1.1 Analyzing total revenue growth over time.

```python
daily_sales = df.groupby("Date").agg({"Sales": "sum"}).reset_index()
daily_sales = daily_sales.rename(columns={"Sales": "Total Revenue"})
daily_sales = daily_sales.sort_values("Date")
daily_sales["Previous_Day_Sales"] = daily_sales["Total Revenue"].shift(1)
daily_sales["Daily_Sales_Growth"] = ((daily_sales["Total Revenue"] - daily_sales["Previous_Day_Sales"]) /daily_sales["Previous_Day_Sales"]) * 100

daily_sales["Daily_Sales_Growth"].fillna(0, inplace=True)
daily_sales["Daily_Sales_Growth"].replace([float("inf"), -float("inf")], 0, inplace=True)
daily_sales
```

1.2 Identifying best-selling products and seasonal sales trends.

```python
daily_category_sales = df.groupby(["Month", "Category"])["Sales"].sum().reset_index()
most_selling_daily = daily_category_sales.sort_values(['Month', 'Sales'], ascending=[True, False]).groupby("Month").first().reset_index()

most_selling_daily
```

1.3 Calculating the average order value to understand spending behavior.

```python
state_summary = (df.groupby("ship-state").agg(Orders=("Order ID", "nunique"), Revenue=("Sales", "sum")).assign(AOV=lambda x: x.Revenue / x.Orders).round(2))
state_summary = state_summary.sort_values(["Revenue", "AOV"], ascending=False).reset_index()
state_summary
```

4.2 Analyzing the average quantity ordered for efficient stock management.

```python
AOQ= df.groupby("Category").agg(Total_quantity=("Qty", "sum"), Total_orders=("Order ID", "nunique")).reset_index()
AOQ["Average_Quantity_per_Order"] = AOQ["Total_quantity"] / AOQ["Total_orders"]
AOQ.sort_values("Average_Quantity_per_Order", ascending=False)
AOQ
```

4.3 Enhancing inventory control to prevent stockouts or overstocking.

```python
inventory_control = (df.groupby("Category").agg(Total_quantity=("Qty", "sum"), Total_orders=("Order ID", "nunique")).reset_index())
inventory_control["AOQ_per_Order"] = (inventory_control["Total_quantity"] / inventory_control["Total_orders"])
returned = df[df["Sub-Status"].isin(["Returning to Seller", "Rejected by Buyer", "Returned to Seller"])]
returned_orders = returned.groupby("Category")["Order ID"].nunique().reset_index(name="Returned_Orders")
inventory_control = inventory_control.merge(returned_orders, on="Category", how="left")
inventory_control["Returned_Orders"] = inventory_control["Returned_Orders"].fillna(0)
inventory_control["Return_Rate"] = ((inventory_control["Returned_Orders"] / inventory_control["Total_orders"]) * 100).round(3)
inventory_control["Stock_Out_Prevention"] = (inventory_control["Total_quantity"] * 0.05).round(2)
inventory_control = inventory_control.sort_values("Return_Rate", ascending=False)
inventory_control
```

#**CUSTOMER SATISFACTION & RETURNS**

5.1 Investigating high return and cancellation rates.

```python
df["Sub-Status"].unique()
```

```python
df["Returned"] = df["Sub-Status"].isin(["Returning to Seller", "Rejected by Buyer", "Returned to Seller"])
df["Cancelled"] = df["Courier Status"] == "Cancelled"
category_summary = (df.groupby("Category").agg(Total_Orders=("Order ID", "nunique"),Returned_Orders=("Returned", "sum"),Cancelled_Orders=("Cancelled", "sum")).reset_index())
category_summary["Return_Rate (%)"] = ((category_summary["Returned_Orders"] / category_summary["Total_Orders"]) * 100).round(2)
category_summary["Cancel_Rate (%)"] = ((category_summary["Cancelled_Orders"] / category_summary["Total_Orders"]) * 100).round(2)
category_summary = category_summary.sort_values(["Cancel_Rate (%)", "Return_Rate (%)"],ascending=False)
category_summary
```

```python
plt.figure(figsize=(12, 7))
plt.barh(category_summary['Category'], category_summary['Cancel_Rate (%)'], color='orange', label='Cancel Rate')
plt.barh(category_summary['Category'], category_summary['Return_Rate (%)'], left=category_summary['Cancel_Rate (%)'], color='red', label='Return Rate')
plt.xlabel('Rate (%)')
plt.title('Top 10 Categories by Cancel and Return Rates')
plt.legend()
plt.show()
```

5.2 Identifying problematic products/categories to enhance quality.

```python
df["Sub-Status"].unique()
```

```python
#if return rate > 15% and total order > 10 its problematic
high_return_products = (
    df.assign(Returned=df["Sub-Status"].isin(["Returning to Seller", "Rejected by Buyer", "Returned to Seller"]))
      .groupby(["Category", "SKU"]).agg(Total_Revenue=("Sales", "sum"),
      Total_Orders=("Order ID", "nunique"),Returned_Orders=("Returned", "sum"))
      .assign(Return_Rate=lambda x: (x["Returned_Orders"] / x["Total_Orders"] * 100).round(2))
      .query("Total_Orders > 10 and Return_Rate > 15").reset_index())
high_return_products
```

B2B vs B2C: B2C sales is actually makes the complete sales as per the conclusion. but there is a significant gap between the AOV between the respective buyers which should to noticed to increase in overall revenue.


Recommendation: There is a significant gap between the AOV between the B2C and B2B which should be noticed to target the sales campaign for the increase in sales.

Average Ordered Quantity:

The analysis has uncovered both strengths and areas of improvement across the business.
